MATPLOTLIB Tutorial

```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
"""Use Numpy to generate randdom data """
import numpy as np

x = np.linspace(0, 10, 25)  #generate nums btw 0 & 10 and use 25 of them.
y = x * x + 2
print(x)
print(y)
```

```python
"""It only takes 1 command to draw our graph"""
pylab.plot (x,y, 'r')  # r-stands for red : b, g, y, etc. (p for a dotted line)
```

```python
 """drawing a subgraph 
"""
#Contents(row, columns, indexes)
#colour and line style

pylab.subplot(1,2, 1) 
pylab.plot(x,y, 'r--') 
pylab.subplot(1,2, 2)
pylab.plot(x,y, 'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5, 0.1, 0.8, 0.8]) #Control the left, right, width, height of the canvas (from 0 to 1)

axis.plot(x,y, 'r')
```

```python
#drawing subgraphs
fig, axes = plt.subplots(nrows = 1, ncols=2) #submap is of 1 row, 2 columns
for ax in axes:
    ax.plot(x,y, 'b')
```

```python
# we can also draw a picture, or a graph inside another graph

fig = plt.figure()

#Control the left, bottom, width, height of the canvas(from 0 to 1)

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #big axes
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2]) #small canvas

axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
```

```python
fig = plt.figure(figsize=(16,9), dpi=300) #New graphic object #dpi-dotsPerInch

fig.add_subplot()

plt.plot(x,y, 'r')
```

```python
fig, axes = plt.subplots()
axes.set_xlabel("x-label")
axes.set_ylabel("y-label")

axes.set_title("title of graph")

axes.plot(x,y, 'r')
```

```python
ax.legend(["label1", "label2"])


fig, axes = plt.subplots()
axes.set_xlabel("x-label")
axes.set_ylabel("y-label")

axes.set_title("title of graph")

axes.plot(x, x**2)
axes.plot(x, x**3)

axes.legend(["y = x**2", "y = x**3"], loc=2)
```

```python
#In Matplotlib, you can set other properties such as line colour, transparency, and more.


fig, axes = plt.subplots(dpi=150)
axes.plot(x, x+1, color='red', alpha=.5)
axes.plot(x, x**2, color='green', alpha=.5)
axes.plot(x, x+3, alpha=.5)
```

```python
# line width and linestyle
fig, axes = plt.subplots(dpi=150)
axes.plot(x, x+1, color='red', lw=3, linestyle=':')
axes.plot(x, x+2, color='green', lw=1, linestyle='-.')
axes.plot(x, x+3, color='black', lw=2, linestyle='-')
```

```python
#set a marker type
fig, axes = plt.subplots(dpi=150)
axes.plot(x, x+1, color='red', marker='o')
axes.plot(x, x+2, color='green', marker='s')
axes.plot(x, x+3, color='black', marker='3')
axes.plot(x, x+4, color='black', marker='+', markersize=8, markerfacecolor='green')
```

```python
""" Set the canvas grid and axis range
"""

fig, axes = plt.subplots(1, 2, figsize=(10,5))
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].set_ylim([0,60])
axes[1].set_xlim([2,5])
axes[1].grid(True)
```

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(16,5))
axes[0].set_title("scatter")
axes[0].set_xlabel("x-axis")
axes[0].set_ylabel("y-axis")
axes[0].scatter(x, x + 0.25*np.random.randn(len(x)))

axes[1].set_title("step")
axes[1].step(n, n**2, lw=2)

axes[2].set_title("bar")
axes[2].bar(n, n**2, align="center", color="green")

axes[3].set_title("fill_between")
axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5)
```

```python
"""Draw a radar chart
"""

fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0, 0.0, 1, 1], polar=True)
t = np.linspace(0, 2 * np.pi, 100)
ax.plot(t, t*2, color='blue', lw=3) 
```

```python
"""Draw a histogram
"""
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))
axes[0].set_title("Default histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative detailed histogram")
axes[1].hist(n, cumulative=True, bins=10)
```

```python
# Draw contour image

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 -  Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1-Z2) * 2

print(X)
print(Y)
print(Z)
```

```python
fig, axes = plt.subplots()
CS = axes.contour(X, Y, Z)
axes.clabel(CS, inline=1, fontsize=10)
axes.set_title("contour map")
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize=(14,6))

#pecify the 3D graphics to draw, with projection= '3d'
axes = fig.add_subplot(1,2,1,projection='3d')
axes.plot_surface(X,Y,Z, rstride=4, cstride=4, linewidth=0)
```

```python
import matplotlib.pyplot as plt
# Data to plot

languages = ['Python', 'Java', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"]
```

```python
#explode 1st slide

explode = (0.1,0,0,0,0,0.2)
plt.pie(popularity, explode=explode, labels=languages, colors=colors, autopct='%1.1f%%', 
        shadow=True, startangle=90)
```
