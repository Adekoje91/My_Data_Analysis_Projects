#  EXPLORATORY DATA ANALYSIS 

#  what is the overall sales trend?

```python
df["order_date"].min()
```

```python
df["order_date"].max()
```

```python
# Getting just month&year from the dataset

df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
df['month_year']
```

```python
# Grouping month year
df_monthly_trend = df.groupby('month_year').sum()['sales'].reset_index()
```

```python
plt.figure(figsize=(15,6))

plt.plot(df_monthly_trend['month_year'], df_monthly_trend['sales'], color='#b80045')

# rotating the x-display to show vertically
plt.xticks(rotation='vertical', size=8)
plt.show()
```

<h4> * WHAT ARE THE TOP 10 PRODUCTS BY SALES?

```python
# Grouping Product name as column
#groupby function is just like a pivot-table, the column you are grouping by becomes the pivot.
#the sales was the only column selected among the list of columns.

prod_sales = pd.DataFrame(df.groupby("product_name").sum()["sales"])
prod_sales
```

```python
prod_sales = prod_sales.sort_values("sales", ascending=False)
```

```python
#Top 10 products by sales
# you can equally use the dictionary [:10]
prod_sales.head(10)
```

<h4> * MOST SELLING PRODUCTS

```python
Most_sell_prod = pd.DataFrame(df.groupby("product_name").sum()["quantity"])
Most_sell_prod
```

```python
Most_sell_prod = Most_sell_prod.sort_values("quantity", ascending=False)
Most_sell_prod[:10]
```

<h4> *WHICH ARE THE MOST PROFITABLE CATEGORY AND SUB-CATEGORY?

```python
# Most profitable category and sub-category
#Coz we have to group more than one column we are using an array brakets []

profitt = pd.DataFrame(df.groupby(["category", "sub_category"]).sum()["profit"])
profitt
```

```python
# I am sorting by profit alone coz it highlights the huge profits across each of the categories.

#example sorted using the category and profit columns
profitt.sort_values("profit", ascending=False)
```
